```python
from urllib import request
from bs4 import BeautifulSoup
```

```python
wiki_page = 'https://en.wikipedia.org/wiki/List_of_songs_recorded_by_Green_Day'
```

```python
request_page = request.urlopen(wiki_page)
page = request_page.read()
page = page.decode('utf8')
soup = BeautifulSoup(page, 'html.parser')
```

```python
td_tags = soup.find_all('td')
```

```python
Song_Titles = []
for tag in td_tags:
    if len(tag.text) > 0 :
        if '"' == tag.text[0]:
            Song_Titles.append(tag.text.replace('"',''))
```
